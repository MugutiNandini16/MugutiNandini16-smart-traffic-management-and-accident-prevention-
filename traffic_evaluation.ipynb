```python
import cv2
import matplotlib.pyplot as plt

video_path = f"{path}/sample_video.mp4"
cap = cv2.VideoCapture(video_path)

width = int(cap.get(3))
height = int(cap.get(4))
fps = int(cap.get(5))

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter("output_video.mp4", fourcc, fps, (width, height))

vehicle_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    for result in results:
        vehicle_count += len(result.boxes)

    for result in results:
        frame = result.plot()

    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    out.write(frame)

cap.release()
out.release()

```

```python
vehicle_count = 0

for result in results:
    detections = result.boxes
    vehicle_count += len(detections)

print(f"Total Vehicles Detected: {vehicle_count}")
```

```python
from ultralytics import YOLO

# Load the trained YOLOv8 model (update path if needed)
model = YOLO('runs/detect/train/weights/best.pt')

# Evaluate on validation set
metrics = model.val()

# Print evaluation metrics
print("Precision:", metrics.box.precision)
print("Recall:", metrics.box.recall)
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

results_dir = 'runs/detect/train'
results_csv = os.path.join(results_dir, 'results.csv')

# Plot training loss if CSV exists
if os.path.exists(results_csv):
    df = pd.read_csv(results_csv)
    df[['      epoch', 'box_loss', 'cls_loss', 'dfl_loss']].plot(figsize=(10, 6), title='Loss During Training')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
else:
    print("results.csv not found. Please ensure training results exist.")
```
